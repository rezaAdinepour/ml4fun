```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../Datasets/melb_data.csv")

print(f"df shape: {df.shape}")
df.head(10)
```

```python
df_cp = df.copy()

df_cp = df_cp.dropna(axis=0)

print(f"df shape: {df_cp.shape}")
df_cp.head(10)
```

```python
features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']

X = df[features]
y = df["Price"]
```

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, train_size=0.8, shuffle=True)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(random_state=1)
model.fit(train_X, train_y)
```

```python
from sklearn.metrics import mean_absolute_error

pred_y = model.predict(val_X)
print(f"MAE: {mean_absolute_error(val_y, pred_y)}")
```

```python
point = np.arange(2, 50, 5)

pred_y_arr = []
MAEs = []

for idx, i in enumerate(point):
    model = RandomForestRegressor(random_state=i)
    model.fit(train_X, train_y)
    pred_y_arr.append(model.predict(val_X))
    MAEs.append(mean_absolute_error(pred_y_arr[-1], val_y))
    print(f"point {i}: MAE: {MAEs[-1]}")
```
