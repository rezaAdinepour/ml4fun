```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Data/non-linear-regression-x-y.csv")

print(f"shape of df is: {df.shape}")
df.head(10)
```

```python
plt.plot(df["x"], df["y"])
plt.title("Time series")
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

```python
model = np.poly1d(np.polyfit(df["x"], df["y"], 50))
```

```python
t = np.linspace(0, 250, 1000)
```

```python
plt.plot(df["x"], df["y"], label="True")
plt.plot(t, model(t), label="Fit")
plt.legend()
```

```python
point = np.array([1, 5, 10, 15, 20, 25, 30, 35])

plt.figure(figsize=(20, 20))

for idx, val in enumerate(point):
    model = np.poly1d(np.polyfit(df["x"], df["y"], val))
    plt.subplot(4, 2, idx + 1)
    plt.title(f"{val} point")
    plt.plot(df["x"], df["y"], label="True")
    plt.xlabel("sample")
    plt.ylabel("Amplitude")
    plt.plot(t, model(t), label="fit")
    plt.legend()
```
